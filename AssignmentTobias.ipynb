# **2.Filtering DataFrame:**

Filtering a dataframe is a method used to subset the columns or rows according to their label or specified index. This is commonly used if a boolean column is present. However you can create your own function.


```
name = apps['App']

#Creating the condition to filter
best = apps['Rating'] > 4.5

application = apps[best]

#displaying appications with the condition of a rating that is greater that 4.5
display(application)
```

# **3.Filtering a DataFrame using Numpy Operation:**

Numpy means numerical python which is a libary that can be imported into your python dataframe. It allows the user to use logical operations efficiently sort through their dataset.

```
#We define rtg as a series because it is a column
rtg = apps['Rating']

#we must import the numpy  libary
import numpy as np

#allows us to display game rating between the two values we input 
between = np.logical_and(rtg > 3, rtg < 4.8)

#I assume that the game rating average between 3 and 4.8
average = apps[between]

display(average)
```

# **4.Looping over the DataFrame:**

This is a method also known as iteration that allows the user to go through a dataset row by row to print out the label or column.

```
#I had to limit it to iterate over only the first 20 because there are thousands of apps recorded in this dataset
for label, row in apps.head(20).iterrows() :
    display(row)
    display(label)
```

# **5.Adding Column:**

This method is used to add a column to your dataset and can be achieved using two methods.
*   Deep Copy
*   Shallow Copy

> Deep copy also known as full copy is the recomended way because it allows you to change something in the copied verison without it affecting the original version.

> Shallow Copy means that if you change something in the original it will also be changed in the copied version.

```
apps_copy = apps.copy(deep = True) #this is a deep copy

for label, row in apps_copy.head(10).iterrows():
    apps_copy.loc[label,'GENRES'] = row['Genres'].upper() #Changes the name values of apps to lower case

display(apps_copy)

apps_copy['genre_small'] = apps_copy['Genres'].apply(str.lower)

display(apps_copy)

display(apps.head(10))
```

# **7.Lambda Function:**

This is a function that can be applied to both rows and columns of a dataframe.
If the axis of the apply function is 0 then lambda is applied to each column and if the apply function is 1 then the lambda is applied to the row.

Referenced from www.geeksforgeeks.org

```
apps_copy = apps.copy(deep = True) #this is a deep copy

#adding 0.2 to each of the app ratings
a = lambda x: x + 0.2

apps_copy['new_Rating'] = apps_copy['Rating'].apply(a)

display(apps_copy.head(10))
```

# **8.Minimun/Maximum:**

This is a method used to return the higest or lowest numerical value in a specified column.

```
apps_copy = apps.copy(deep = True) #this is a deep copy

#finding the highest ranked game
apps_copy.max(axis = 1)
```

# **9.Conditional Changes:**

This is a method used to change values or names within the dataframe based on the conditions used to filter the output of the method.

```
apps_copy = apps.copy(deep = True) #this is a deep copy

#Before the change
print('Before the change')
apps_copy.head(10)

#Changing the name of the category ARTS_AND_DESIGN to Arts & Crafts
print('After the change')
apps_copy.loc[apps_copy['Category'] == 'ART_AND_DESIGN', 'Category'] = 'Arts & Crafts'

apps_copy
```

# **10.Aggregate Statistics using Groupby:**

This function allows you to do more advanced analysis on the dataframe. After grouping the usr can pass aggregation functions to the grouped object as a dictionary within the aggregated function.

```
apps_copy = apps.copy(deep = True) #this is a deep copy

apps_copy.groupby(['App']).mean().sort_values('Rating')
```

**Method to find the sum (Part of 10)**



```
apps_copy = apps.copy(deep = True) #this is a deep copy

apps_copy.groupby(['App']).sum().sort_values('Rating')
```
